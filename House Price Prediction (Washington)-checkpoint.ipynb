## Train - Test split

```python
# Train test split
X = dummy_df.drop('price' , axis = 1)
Y = dummy_df['price']
```

```python
# Standardizing numerical features using standard scalar
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)
```

```python
# Creating function for different regression metrics
def regression_metrics(Y_test, Y_pred, no_of_input_features = None):
    
    # Calculate the root mean squared error
    rmse = mean_squared_error(Y_test, Y_pred, squared=False)
    print(f"RMSE: {rmse:.3f}\n")

    # Calculate the adjusted R-squared
    r2 = r2_score(Y_test, Y_pred)
    adj_r2 = 1 - ((1 - r2) * (len(Y_test) - 1) / (len(Y_test) - no_of_input_features - 1))  if no_of_input_features is not None else r2
    print(f"Adj_R2: {adj_r2:.3f}\n") if no_of_input_features is not None else print(f"R2: {r2:.3f}\n")

    # Calculate the mean absolute percentage error
    mape = (1 / len(Y_test)) * np.sum(np.abs((Y_test - Y_pred) / Y_test)) * 100
    print(f"MAPE: {mape:.3f}\n")

    # Calculate the explained variance score
    evs = explained_variance_score(Y_test, Y_pred)
    print(f"EVS: {evs:.3f}\n")

    # Calculate the maximum error
    max_err = max_error(Y_test, Y_pred)
    print(f"Max_Error: {max_err:.3f}\n")
    
    return rmse , adj_r2 if no_of_input_features is not None else r2 , mape , evs , max_err
```

## Linear Regression

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
Y_pred_lr = lr.predict(X_test)
regression_metrics(Y_test,Y_pred_lr,14)
```

## StackedCVRegression Method

### Training base models

```python
lgbr = LGBMRegressor(objective='regression',num_leaves=5,
                              learning_rate=0.05, n_estimators=720,
                              max_bin = 55, bagging_fraction = 0.8,
                              bagging_freq = 5, feature_fraction = 0.2319,
                              feature_fraction_seed=9, bagging_seed=9,
                              min_data_in_leaf =6, min_sum_hessian_in_leaf = 11)

hgbr = HistGradientBoostingRegressor(learning_rate=0.05,max_depth=4,
                                     min_samples_leaf=15,random_state =5)

gbr = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber', random_state =5)

xgbr = XGBRegressor(colsample_bytree=0.4603, gamma=0.0468, 
                             learning_rate=0.05, max_depth=3, 
                             min_child_weight=1.7817, n_estimators=2200,
                             reg_alpha=0.4640, reg_lambda=0.8571,
                             subsample=0.5213, silent=1,
                             random_state =7, nthread = -1)

rfr = RandomForestRegressor()
```

```python
stacked_model = StackingCVRegressor(regressors=[lgbr , hgbr , gbr , xgbr , rfr], meta_regressor= lgbr , cv = 5)
```

```python
stacked_model.fit(X_train, Y_train)
```

```python
stacked_Y_preds = stacked_model.predict(X_test)
```

```python
regression_metrics(Y_test , stacked_Y_preds , 14)
```
